```python
import pandas as pd
import os
import glob
```

```python
path = '../data/cleaned'
files = glob.glob(os.path.join(path, '*.pkl'))
data = {os.path.basename(f)[:-4]: pd.read_pickle(f) for f in files}
```

```python
full_dataset = data['openweather']\
    .merge(data['government'], how='outer', left_index=True, right_index=True)\
    .merge(data['satellite'], how='outer', left_index=True, right_index=True).sort_index()
```

```python
full_dataset.to_csv(os.path.join(path, 'full_dataset.csv'))
full_dataset.to_pickle(os.path.join(path, 'full_dataset.pkl'))
full_dataset
```
