```python
import pandas as pd
import os
import glob
```

```python
path = '../data/raw/api_data_IN/'
files = glob.glob(os.path.join(path, '*.csv'))
data = []
for f in files:
    df = pd.read_csv(f).rename(columns={'City': 'location', 'Date': 'date'})
    df['date'] = pd.to_datetime(df['date'])
    df = df.set_index(['date', 'location'])[['CO', 'NO2', 'SO2', 'PM2_5']]
    data.append(df)
```

```python
final = pd.concat(data)
final.columns = [f + '_openweather' for f in final.columns]
final = final.astype('float64')
final.to_csv('../data/cleaned/openweather.csv')
final.to_pickle('../data/cleaned/openweather.pkl')
final
```
